## Multi Class Logistic Regression 

### Load Libraries 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Load Data 

```python
iris = load_iris()
```

```python
X,y = iris.data,iris.target
```

### Scale the Data 

Always make sure to scale the data before running Logistic Regression as it involves Gradient Descent and it can be computantionally expensive specially if data size is huge and its not scaled.

```python
# Scale the data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X
```

```python
y
```

```python
X_scaled
```

### Split the Data in Train and Test Set 

```python
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.33, random_state=42)
```

### Train the Model 

```python
log_model = LogisticRegression(random_state=0, multi_class='ovr')
```

```python
log_model.fit(X_train,y_train)
```

### Test the Model 

```python
pred = log_model.predict(X_test)
```

```python
pred
```

### Check Accuracy of the Model on Test Data 

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
accuracy_score(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```

### Predict on Completely New Data 

```python
log_model.predict(scaler.transform([[5.9, 3. , 5.1, 1.8]]))
```
